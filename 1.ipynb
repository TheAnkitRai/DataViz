```python
from bokeh.layouts import column
from bokeh.models import ColumnDataSource, DataTable, TableColumn
from bokeh.plotting import figure, show, output_notebook
import pandas as pd
```

```python
output_notebook()
```

```python
from collections import namedtuple
Color = namedtuple('RGB', 'red, green, blue')
class RGB(Color):
    def hex_format(self):
        return '#{:02X}{:02X}{:02X}'.format(self.red, self.green, self.blue)
```

```python
data = pd.read_csv('Q1_data.csv')
source = ColumnDataSource(data = {
    'x': data['PassengerId'],
    'y': data['Survived']
})
```

```python
columns = [
    TableColumn(field = 'x', title = 'PassengerID'),
    TableColumn(field = 'y', title = 'Survived')
]
```

```python
data_table = DataTable(source = source, columns = columns, editable = True, width = 800, index_position = -1, index_header = 'Row Index', index_width = 60)
```

```python
p = figure(width = 800, height = 300, tools = 'pan, wheel_zoom, xbox_select, lasso_select, reset', active_drag = 'xbox_select')
```

```python
colors = [RGB(255, 0, 0) if value == 0 else RGB(0, 255, 0) for value in source.data['y']]
#p.circle(x = 'x', y = 'y', source = source, fill_color = colors)
p.circle(x = 'x', y = 'y', source = source, fill_color = 'red')
```

```python
show(column(p, data_table))
```
